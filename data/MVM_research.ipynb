# 2. Предварительная обработка данных

```python
#Удалим customerID
data.drop(['customerID'], axis = 1, inplace = True)
```

```python
data
```

## 2.1. Заменим признаки на их бинарное представление


```python
data.gender = data.gender.apply(lambda x: 0 if x == 'Female' else 1)
data.Partner = data.Partner.apply(lambda x: 0 if x == 'No' else 1)
data.Dependents = data.Dependents.apply(lambda x: 0 if x == 'No' else 1)
data.PhoneService = data.PhoneService.apply(lambda x: 0 if x == 'No' else 1)

#Заменим No phone service на No в MultipleLines
data['MultipleLines'] = data['MultipleLines'].replace('No phone service','No')
#Заменим No internet service на No
data['OnlineSecurity'] = data['OnlineSecurity'].replace('No internet service','No')
data['DeviceProtection'] = data['DeviceProtection'].replace('No internet service','No')
data['TechSupport'] = data['TechSupport'].replace('No internet service','No')
data['StreamingTV'] = data['StreamingTV'].replace('No internet service','No')
data['StreamingMovies'] = data['StreamingMovies'].replace('No internet service','No')


data.MultipleLines = data.MultipleLines.apply(lambda x: 0 if x == 'No' else 1)
data.OnlineSecurity = data.OnlineSecurity.apply(lambda x: 0 if x == 'No' else 1)
data.DeviceProtection = data.DeviceProtection.apply(lambda x: 0 if x == 'No' else 1)
data.TechSupport = data.TechSupport.apply(lambda x: 0 if x == 'No' else 1)
data.StreamingTV = data.StreamingTV.apply(lambda x: 0 if x == 'No' else 1)
data.StreamingMovies = data.StreamingMovies.apply(lambda x: 0 if x == 'No' else 1)
data.Churn = data.Churn.apply(lambda x: 0 if x == 'No' else 1)

#Удалим строки с пустыми значениями TotalCharges 
data['TotalCharges'] = data['TotalCharges'].replace(' ', None)
data.dropna(subset=['TotalCharges'], inplace=True)
#Конвертируем в float
data["TotalCharges"] = data["TotalCharges"].astype(float)
X_columns = ['gender']#["InternetService", "PaymentMethod", ]
X_categ = data[X_columns]
```
